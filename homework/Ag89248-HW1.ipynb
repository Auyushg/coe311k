```python
import matplotlib.pyplot as plt
import numpy as np
import math
import random

def monteCarlo(numSamples):
    counter = 0
    x=0
    y=0
    for s in range(0,numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if((x**2 + y**2) < 1):
            counter = counter + 1
    return counter/numSamples * 4
    
```

```python
pi = np.pi
x = np.arange(10,10000,20)
y = []
for i in range(len(x)):
    y.append(monteCarlo(x[i]))
y = np.array(y)
y = np.abs(y - pi)
plt.plot(x,y,'r--')
plt.show()
```
