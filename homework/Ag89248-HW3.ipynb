```python
import numpy as np
import matplotlib.pyplot as plt

def factorial(n):
    if (n==1):
        return 1
    return n * factorial(n-1)



def sin_taylor(numTerms, x):
    estimation = 0
    for n in range(numTerms):
        estimation += ((-1**n)/factorial(2 * n + 1)) * x**(2*n+1)
    return estimation
        
```

```python
x = np.linspace(0,10,101)
sin1 = []
sin2 = []
sin3 = []
sin4 = []
sine = np.sin(x)
for i in range(len(x)):
    sin1.append(sin_taylor(1,x[i]))
    sin2.append(sin_taylor(2,x[i]))
    sin3.append(sin_taylor(3,x[i]))
    sin4.append(sin_taylor(4,x[i]))
sin1 = np.array(sin1)
sin2 = np.array(sin2)
sin3 = np.array(sin3)
sin4 = np.array(sin4)

t1 = np.abs(sine-sin1)
t2 = np.abs(sine-sin2)
t3 = np.abs(sine-sin3)
t4 = np.abs(sine-sin4)

plt.plot(x,t1, 'r')
plt.plot(x,t2, 'g')
plt.plot(x,t3, 'b')
plt.plot(x,t4, 'p')
```
