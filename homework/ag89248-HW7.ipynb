```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline


# Define parameters
h = 0.01  # step size
h_list = []
error_list = []
x = 15
k = 1
error_x15_list = []
l = -0.05

while h<=1:
    x_start, x_end = 0, 10 * np.pi
    num_steps = int((x_end - x_start) / h)
    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)

    # Initial condition
    y_values[0] = 1  # y(0) = 1
    # Function representing dy/dx = -2y
    def f(y):
        return l * y

    # Implement Euler's method
    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * f(y_values[i])
    h_list.append(h)
    
    y_avg = np.mean(y_values)
    if(h==0.01):
        y_init = y_avg
    error = np.abs(y_avg - y_init)
    error_list.append(error)
    
    difference_array = np.absolute(x_values-x)
    index = difference_array.argmin()
    if(h==0.01):
        val_init = y_values[index]
    errorx = np.abs(val_init - y_values[index])
    error_x15_list.append(errorx)
    
    is_stable = np.abs(1 + h * l) < 1
    print(f"Stabillity for h={h} and l={l}: {is_stable}")
    
    

    # Plot the results
    plt.plot(x_values, y_values, label=f"Euler's Method Approximation, h={round(h, 2)}")
    if(h==0.01):
        h+=0.04
    else:
        h +=0.05
    
```
