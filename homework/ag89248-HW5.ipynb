```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
def my_lstsq(X, Y):
    n = len(X)
    sumX = np.sum(X)
    sumY = np.sum(Y)
    sumx2 = np.sum(X**2)
    sumA = np.sum(X * Y)
    m = (n * sumA - sumX * sumY) / (n * sumx2 - sumX**2)
    c = (sumY - m * sumX) / n
    return m, c
```

let m = 3.14159 and c = 2.71

Testing random matrices:

```python
X = np.sort(random.rand(10))
Y = random.rand(10)
m, c = my_lstsq(X, Y)
Y2 = X * m + c
plt.scatter(X, Y, label="Random Points")
plt.plot(X, Y2, label = "LSTSQ Estimate")
plt.legend()
```

```python
m = 3.14159
c = 2.71
X = np.linspace(0, 1, 101)
p = random.rand(len(X)) * 2 - 1
Y = m * X + c + p
predm, predc = my_lstsq(X, Y)
Y2 = X * predm + predc
plt.scatter(X,Y, label="Points on actual curve")
plt.plot(X, Y2, label="LSTSQ Estimate")
print(f"Predicted m value: {predm}")
print(f"Predicted c value: {predc}")
yE = np.abs(Y2 - Y)
maxYE = np.max(yE)
i = np.argwhere(yE == maxYE)[0]
maxXE = X[i]
print(f"Residuals:{p}")
print(f"Max residual error at X={maxXE}, Y-difference={maxYE}")
minYE = np.min(yE)
i = np.argwhere(yE == minYE)[0]
minXE = X[i]
print(f"Min residual error at X={minXE}, Y-difference={minYE}")
A = np.vstack([X, np.ones(len(X))]).T
# Solve using least squares
m2, c2 = np.linalg.lstsq(A, Y, rcond=None)[0]
Y3 = X * m2 + c2
plt.plot(X, Y3, "r--", label="Numpy LSTSQ Estimate")
plt.legend()
```
