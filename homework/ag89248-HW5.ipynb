```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
def my_lstsq(X, Y):
    n = len(X)
    sumX = np.sum(X)
    sumY = np.sum(Y)
    sumx2 = np.sum(X**2)
    sumA = np.sum(X * Y)
    m = (n * sumA - sumX * sumY) / (n * sumx2 - sumX**2)
    c = (sumY - m * sumX) / n
    return m, c
```
