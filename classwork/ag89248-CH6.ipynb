Jacobi Method

```python
import numpy as np
import time as t
# Jacobi Method
def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
A = np.array([[10, -1, 2, 0, 0],
[-1, 11, -1, 3, 0],
[2, -1, 10, -1, 0],
[0, 3, -1, 8, -2],
[0, 0, 0, -2, 9]])
b = np.array([14, 30, 26, 25, 37])
x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100
time1 = t.time()
solution, iterations = jacobi(A, b, x0, tol, max_iterations)
time2 = t.time()
timeo1 = time2 - time1
print(f"Solution: {solution}")
print(f"Iterations: {iterations}")
print(f"Total Time: {timeo1} seconds")
```

```python
def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
A = np.array([[10, -1, 2, 0, 0],
[-1, 11, -1, 3, 0],
[2, -1, 10, -1, 0],
[0, 3, -1, 8, -2],
[0, 0, 0, -2, 9]])
b = np.array([14, 30, 26, 25, 37])
x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100
time1 = t.time()
solution, iterations = gauss_seidel(A, b, x0, tol, max_iterations)
time2 = t.time()
timeo2 = time2 - time1
print(f"Solution: {solution}")
print(f"Iterations: {iterations}")
print(f"Total Time: {time} seconds")


print(f"The gauss-seidel method is faster by {timeo1 - timeo2} seconds")
```
