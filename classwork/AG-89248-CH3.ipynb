```python
import matplotlib.pyplot as plt
import numpy as np
import math
import random

def monteCarlo(numSamples):
    counter = 0
    x=0
    y=0
    for s in range(0,numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if((x**2 + y**2) < 1):
            counter = counter + 1
    return counter/numSamples * 4
    

y = []

for i in range(100):
    y.append(monteCarlo(1000))
y = np.array(y)
absError = np.abs(np.pi-y)
relError = absError / np.pi
x = np.linspace(1,100, 100)
plt.plot(x,absError, x, relError)
```
