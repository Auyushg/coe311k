```python
import numpy as np
import matplotlib.pyplot as plt
# Define the parameters
k = 1.0        # Constant in the equation (e.g., spring constant)
h = [0.01, 0.02, 0.03, 0.05, .1, .3, .5]       # Step size
x_start = 0    # Start of the interval
x_end = 10     # End of the interval
y_0 = 1.0      # Initial condition for y
v_0 = 0.0      # Initial condition for dy/dx
# Create arrays to store x, y, and v values
plt.figure(figsize=(10, 5))
for i in h:
    x_values = np.arange(x_start, x_end + i, i)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))
    # Set initial conditions
    y_values[0] = y_0
    v_values[0] = v_0

    # Euler's Method loop
    for j in range(1, len(x_values)):
        # Current values
        x_n = x_values[j - 1]
        y_n = y_values[j - 1]
        v_n = v_values[j - 1]    
        # Update equations
        y_values[j] = y_n + i * v_n
        v_values[j] = v_n + i * (-k * y_n)
        

    # Plot the results
    plt.plot(x_values, y_values, label=f'Position y(x), h={i}')
    plt.plot(x_values, v_values, label=f'Velocity v(x), h={i}', linestyle='--')
plt.xlabel('x')
plt.ylabel('y and v')
plt.title('Solution of d²y/dx² = -ky using Euler\'s Method')
plt.legend()
plt.grid(True)
plt.show()
```

This becomes very unstable - the difference in function values increase more rapidly than the difference in h
