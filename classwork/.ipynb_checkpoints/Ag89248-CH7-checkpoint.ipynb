```python
import numpy as np
A = np.array([
[0.582745, 0.48, 0.10	, 0.  	, 0.  	], 
 [0.48	, 1.044129, 0.46	, 0.10	, 0.  	],
 [0.10	, 0.46	, 1.10431 , 0.44	, 0.10	],
 [0.  	, 0.10	, 0.44	, 0.963889, 0.42	],
 [0.  	, 0.  	, 0.10	, 0.42	, 0.522565]], dtype=float)
b = np.array([1.162745, 2.084129, 2.20431 , 1.923889, 1.042565], dtype=float)
x0 = np.zeros_like(b)
```

```python
def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations

def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations
```

```python
tol = 1e-6
mi = 100
soln, it1 = jacobi(A, b, x0, tol, mi)
soln2, it2 = gauss_seidel(A, b, x0, tol, mi)
print(f"Jacobi iterations: {it1}")
print(f"Gauss-seidel: {it2}")
print(f"Soln Jacobi: {soln}")
print(f"Soln Gass-Seidel: {soln2}")
```

```python
A = np.array([[1, 	1], [1, 1.0001]])
b = np.array([1, 1])
x0 = np.zeros_like(b)
soln, it1 = jacobi(A, b, x0, tol, mi)
soln2, it2 = gauss_seidel(A, b, x0, tol, mi)
print(f"Jacobi iterations: {it1}")
print(f"Gauss-seidel: {it2}")
print(f"Soln Jacobi: {soln}")
print(f"Soln Gass-Seidel: {soln2}")
```

```python
b = np.array([1, 1.001])
x0 = np.zeros_like(b)
soln, it1 = jacobi(A, b, x0, tol, mi)
soln2, it2 = gauss_seidel(A, b, x0, tol, mi)
print(f"Jacobi iterations: {it1}")
print(f"Gauss-seidel: {it2}")
print(f"Soln Jacobi: {soln}")
print(f"Soln Gass-Seidel: {soln2}")
```

Because some matrices are ill-conditioned(close to being the same), rounding errors can be amplified, and the two methods will slow down or even fail to converge to a solution within the tolerance

```python
A = np.array([
[0.582745, 0.48, 0.10	, 0.  	, 0.  	], 
 [0.48	, 1.044129, 0.46	, 0.10	, 0.  	],
 [0.10	, 0.46	, 1.10431 , 0.44	, 0.10	],
 [0.  	, 0.10	, 0.44	, 0.963889, 0.42	],
 [0.  	, 0.  	, 0.10	, 0.42	, 0.522565]], dtype=float)
b = np.array([1.162745, 2.084129, 2.20431 , 1.923889, 1.042565], dtype=float)
x0 = np.zeros_like(b)

def is_diagonal(A):
    for i in range(len(A)):
        dval = np.abs(A[i][i])
        sum = 0
        for j in range(len(A[0])):
            if(j == i):
                sum = sum + 0
            else:
                sum = sum + np.abs(A[i][j])
        if(dval < sum):
            return False
    return True

print("A:")
print(A)
print(is_diagonal(A))          
```

```python
C = np.array([[1, 2, 3, 0, 0], [2, 1, 2, 3, 0], [3, 2, 1, 2, 3], [0, 3, 2, 1, 2], [0, 0, 3, 2, 1]])
print("A:")
print(C)
print(is_diagonal(C)) 
```
