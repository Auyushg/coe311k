```python
import numpy as np
import time as t
import matplotlib.pyplot as plt
matrices = []
for i in range(8):
    n = 3 * 2**i
    a = np.random.rand(n, n)
    b = np.random.rand(n, n)
    matrices.append(a)
    matrices.append(b)
    
def naive_matrix_multiplication(A, B):
    t1 = t.time()
    n = len(A)
    C = [[0] * n for _ in range(n)]

    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    t2 = t.time()
    dt = t2-t1
    return dt


def optimized(A, B):
    t1 = t.time()
    np.matmul(A, B)
    t2 = t.time()
    dt = t2-t1
    return dt

len_matrix = []
time_naive = []
time_optimized = []

for i in range(0, len(matrices), 2):
    len_matrix.append(len(matrices[i]))
    time_naive.append(naive_matrix_multiplication(matrices[i], matrices[i+1]))
    time_optimized.append(optimized(matrices[i], matrices[i+1]))

print(len(time_naive))
plt.plot(len_matrix, time_naive, label = "unoptimized time")
plt.legend()
plt.show()
plt.plot(len_matrix, time_optimized, label = "optimized time")
plt.legend()
plt.show()
plt.plot(len_matrix, time_naive, label = "unoptimized time")
plt.plot(len_matrix, time_optimized, label = "optimized time")
plt.legend()
plt.show()
```
