```python
import math
import random
def approximatePi(numSamples):
    counter = 0
    x=0
    y=0
    for s in range(0,numSamples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if((x**2 + y**2) < 1):
            counter = counter + 1
    return counter/numSamples * 4
    
    
```

```python
apprximatePi(1000)
```
