For floating points, 32 bits involve 1 for sign, 8 for the exponent, and 23 for the mantissa. In 64 bit, 1 for sign, 11 for exp, and 52 for the mantissa. As the number gets bigger, the prescision drops. 128-bit is typically not used since it reduces performance.

Machine Epsilon: smallest value for which when added to the value the computer recognizes that value has changed. It is equal to 2^t, where t is the number of bits designated to the mantissa. 

Values below machine epsilon neglected by the computer, learing to roundoff errors
