Let's discuss systems of linear equations: take Ax = b, where A is the coefficient matrix, x is the unknown vector, and b is the constant vector

given:
2x + 3y = 5
4x - y = 3

```python
A = np.array([[2,3],[4,-1]])
b = np.array([5,3])
print(A,b)
x = np.linalg.solve(A,b)
print(x)
```

Eigenvalue - given Av - lambda * v, lambda is a scalara and v is a nonzero vector
lambda is your eigenvalue, and v is the eigenvactor

eigenvectors are directions that don't change when a matrix acts on them(directions of maximum variance in data)
eigenvalues control the magnitude of the eigenvectors(how important each direction is)


```python
A = np.array([[2,1],[1,2]])
eigenvalues, eigenvectors = np.linalg.eig(A)
```

```python
print(eigenvalues,eigenvectors)
```

gaussian elim

```python
def gaussian_elimination(A, b):
    n = len(b)
 # Forward elimination
    for i in range(n):
 # Pivoting
        max_row = max(range(i, n), key=lambda x: abs(A[x][i]))
        A[i], A[max_row] = A[max_row], A[i]
        b[i], b[max_row] = b[max_row], b[i]
        for j in range(i + 1, n):
            ratio = A[j][i] / A[i][i]
            A[j] = [A[j][k] - ratio * A[i][k] for k in range(n)]
            b[j] -= ratio * b[i]
    # Backward substitution
    x = [0] * n
    for i in range(n - 1, -1, -1):
        x[i] = (b[i] - sum(A[i][j] * x[j] for j in range(i + 1, n))) / A[i][i]
    return x
```

```python
A = [[3, 2, -4], [2, 3, 3], [5, -3, 1]]
b = [3, 15, 14]
solution = gaussian_elimination(A, b)
print(f"Solution: {solution}")
```

```python
ans = np.linalg.solve(A,b) #Does the EXACT same thing
print(ans)
```

```python
def lu_decomposition(A):
    n = len(A)
    L = np.eye(n) # Initialize L as an identity matrix
    U = np.array(A, dtype=float) # Copy of A to avoid
    # modifying the original matrix
    for i in range(n):
         # Eliminate elements below the pivot (U[i, i])
        for j in range(i+1, n):
            if U[i, i] == 0:
                raise ValueError("Zero pivot encountered. LU decomposition cannot proceed.")
                factor = U[j, i] / U[i, i] # Multiplier for row
                 # operation
                L[j, i] = factor # Store multiplier in L
                 # Update the rows of U
                for k in range(i, n):
                    U[j, k] -= factor * U[i, k]
    return L, U
```

```python
A = np.array([[2, 3, 1],
 [4, 7, -1],
 [-2, 1, 2]])
L, U = lu_decomposition(A)
print("Matrix A:\n", A)
print("Lower Triangular Matrix L:\n", L)
print("Upper Triangular Matrix U:\n", U)
```
