```python
import numpy as np
A = np.array([[2,1], [1,1]])
b = np.array([[2,4], [1,2]])
print(np.linalg.det(A))
print(np.linalg.det(b))
```

```python
a = np.linalg.inv(A)
b = np.linalg.inv(b)
print(np.linalg.det(a))
print(np.linalg.det(b))
```
