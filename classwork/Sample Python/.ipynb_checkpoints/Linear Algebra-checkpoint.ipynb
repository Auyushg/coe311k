Linear Algebra

```python
import numpy as np

v1 = np.array([1,2,3])
v2 = np.array([4,5,6])

print(v1)
print(v2)
```

```python
v_add = v1 + v2
v_sub = v1 - v2

print(v_add)
print(v_sub)

scalar = 5
v_scaled = v1 * scalar
print(v_scaled)
```

```python
dot_product = np.dot(v1,v2)
print(dot_product)
```

```python
#Norm is the magnituude of the vector
magnitude_v1 = np.linalg.norm(v1)
print(magnitude_v1)
```

```python
A = np.array([[1,2,3],[4,5,6]])
print(A)

B = np.array([[7,8,9],[10,11,12],[13,14,15]])
print(B)
```

```python
C = np.array([[1,2,3],[4,5,6]])

matrix_add = A + C
print(matrix_add)
```

```python
matrix_mul = np.dot(A,B)
print(matrix_mul)
```

```python
transpose_a = A.T
print(transpose_a)
```

Let's discuss systems of linear equations: take Ax = b, where A is the coefficient matrix, x is the unknown vector, and b is the constant vector

given:
2x + 3y = 5
4x - y = 3

```python
A = np.array([[2,3],[4,-1]])
b = np.array([5,3])
print(A,b)
x = np.linalg.solve(A,b)
print(x)
```
