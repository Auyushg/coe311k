```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random
# Data points
X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])
# Stack the design matrix with a column of ones (for the intercept)
A = np.vstack([X, np.ones(len(X))]).T
# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(X)
print(Y)
print(A)
print(f"Slope: {m}, Intercept: {c}")
```

```python
plt.scatter(X,Y)
Y2 = X*m + c
plt.plot(X,Y2)
```

```python
X = np.linspace(0, 1, 101)
m = 8.65
c = 8.82
p = random.rand(len(X)) * 2 - 1
Y = m * X + c + p
plt.plot(X, Y)
A = np.vstack([X, np.ones(len(X))]).T
# Solve using least squares
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
Y2 = m * X + c
plt.plot(X, Y2)
print(np.max(p))
print(p)
print(f"Slope: {m}, Intercept: {c}")
yE = np.abs(Y2 - Y)
maxYE = np.max(yE)
i = np.argwhere(yE == maxYE)[0]
maxXE = X[i]
print(f"Max residual error at X={maxXE}, Y-difference={maxYE}")
minYE = np.min(yE)
i = np.argwhere(yE == minYE)[0]
minXE = X[i]
print(f"Min residual error at X={minXE}, Y-difference={minYE}")
```
