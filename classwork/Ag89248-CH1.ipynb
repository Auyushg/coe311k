```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(0,2*np.pi,50)
```

```python
plt.subplot(3,3,1)
plt.plot(x, np.sin(x), 'r-o')
y=np.cos(x)
plt.subplot(3,3,2)
plt.plot(x,y,'g--')

plt.subplot(3,3,3)
plt.hist(x**2,10)
plt.show()

plt.subplot(3,3,4)
n=2000
x= np.random.rand(n)
y=np.random.rand(n)
size = np.random.rand(n) * 50
color = np.random.rand(n)
plt.scatter(x, y, size, color)
plt.colorbar()

plt.show()
```
